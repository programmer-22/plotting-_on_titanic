```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import dataset
kashti = sns.load_dataset("titanic")
kashti
```

```python
# check ko sa parameter kya batha rha hai
kashti.info()
```

```python
# check null values
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind='bar')
```

```python
# another way
sns.countplot(x="survived", hue="sex", data=kashti)
```

```python
import seaborn as sns
```

```python
# scatter plot
plt.scatter(data=kashti, x="age", y="fare", cmap="pclass")# c= Numeric columns for color map
# syntaxc=df["pclass"]
# cmap="pclass"
```

```python
sns.scatterplot(data=kashti, x="age", y="fare", hue="pclass") # Category coloring (seaborn)
#syntax hue="pclass"
```

```python
sns.barplot(data=kashti, x="sex", y="fare", hue="pclass") 
```

```python
# most important plot
sns.pairplot(kashti)# har variable ko other variable kay against draw kya ja tha hai
```

```python
kashti.hist()
```

```python
# hist plot
kashti.hist(bins=10,figsize=(9,7),grid=False)#histogramplot
```

```python
kashti.columns
```

```python
sns.boxplot(data=kashti, x="sex", y="fare", hue="embark_town")
```

```python
sns.boxenplot(data=kashti, x="sex", y="fare", hue="class")
```

# Assignment
How to change the x and y axis titles?

```python
# subplot
g = sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True,
                  palette={1:"seagreen", 0:"gray"})
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True,
                  palette={1:"seagreen", 0:"gray"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend();
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="embark_town", margin_titles=True,
                  palette={1:"seagreen", 0:"gray"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend();
```
